```python
import pandas as pd

# 두 모델의 결과 불러오기
bert = pd.read_csv('/content/drive/MyDrive/LLM/Koelectra-base_seed42.csv')
logistic = pd.read_csv('/content/drive/MyDrive/LLM/kobigbird_seed42.csv')

# Soft Voting 앙상블
ensemble_probs = (bert['generated'] + logistic['generated']) / 2

# 제출 파일 생성
submission = pd.read_csv('/content/drive/MyDrive/LLM/sample_submission.csv')
submission['generated'] = ensemble_probs
submission.to_csv('/content/drive/MyDrive/LLM/koelectra+kobigbird_Ensemble.csv', index=False)
```
